```python


#run  only if you want result about Top_31 and Top_58
Top_31=["human", "man", "woman", "boy", "girl", "father", "mother", "dad", "mom", "grandfather", "grandmother", "baby", "infant", "fetus", "corpse", "dog", "puppy", "cat", "kitten", "frog", "ant", "fish", "mouse", "bird", "shark", "elephant", "beetle", "insect", "chimpanzee", "monkey", "primate"]
               
Top_58 =["human", "man", "woman", "boy", "girl", "father", "mother", "dad", "mom", "grandfather", "grandmother", "baby", "infant", "fetus",   "corpse", "dog", "puppy", "cat", "kitten", "frog", "ant", "fish",  "mouse", "bird", "shark", "elephant", "beetle", "insect", "chimpanzee" , "monkey", "primate", "car", "rock", "hammer", "computer", "robot",  "god", "angle", "ghost", "puppet", "pigeon", "chicken", "rabbit",  "fox", "turkey", "pig", "cow", "horse", "sheep", "lamb", "cucumber",   "lettuce", "potato", "cabbage", "chocolate", "coffee", "tea", "butter"]

delet_index=[]
for i in range(255):
    
     if items['word'].tolist()[i] not in Top_31:
        delet_index.append(i)
        
items=items.drop(delet_index)

```

```python
items['nsubjpass']+=1
items['pobj']+=1
items['dobj']+=1
items['nsubj']+=1


items['(nsubj+nsubjpass)']=items['nsubj']+items['nsubjpass']
items['(nsubj+dobj)']=items['nsubj']+items['dobj']
items['(nsubj+pobj)']=items['nsubj']+items['pobj']
items['(nsubjpass+dobj)']=items['nsubjpass']+items['dobj']
items['(nsubjpass+pobj)']=items['nsubjpass']+items['pobj']
items['(dobj+pobj)']=items['dobj']+items['pobj']



pos_tags_combi= ['nsubj', 'nsubjpass', 'dobj', 'pobj','(dobj+pobj)','(nsubjpass+dobj)','(nsubjpass+pobj)', '(nsubj+pobj)','(nsubj+dobj)','(nsubj+nsubjpass)']

possi_combi_numb=1
for x in pos_tags_combi :
    for y in pos_tags_combi:
        if x != y :  
         items['combination'+str(possi_combi_numb)+x+'/'+y]=items[x]/items[y]
         possi_combi_numb+=1
            
            
            
items['combination(nsubjpass+dobj)/(all)']=((items['nsubjpass'])+(items['dobj']))/(items['dobj']+items['nsubjpass']+items['nsubj']+items['pobj'])
```

```python
correlation_agency=[]
pvalue_agency=[]
correlation_patiency=[]
pvalue_patiency=[]
posib_comb=items.columns[8:].tolist()

from scipy.stats import spearmanr
for x in  items.columns[8:].tolist():
    correlation_agency.append(spearmanr(items['agency'].tolist(),items[x].tolist())[0])
    pvalue_agency.append(spearmanr(items['agency'].tolist(),items[x].tolist())[1]) 
    correlation_patiency.append(spearmanr(items['patiency'].tolist(),items[x].tolist())[0])
    pvalue_patiency.append(spearmanr(items['patiency'].tolist(),items[x].tolist())[1])   
  
```

```python
data = {'posib_comb': posib_comb, 'correlation_agency': correlation_agency,'pvalue_agency':pvalue_agency ,'correlation_patiency':correlation_patiency, 'pvalue_patiency':pvalue_patiency }
spearmanrs= pd.DataFrame(data)

spearmanrs.to_csv('spearmanrs'+'_coha2010_'+'Survey top 31.csv')
```

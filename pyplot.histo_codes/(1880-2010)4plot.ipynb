```python
import numpy as np
import pandas as pd
from zipfile import ZipFile
from matplotlib import pyplot


zip_all_log = ZipFile('plot_log(all_names)1880-2010.zip', 'w')
zip_dec_log= ZipFile('plot_log(decade_names)1880-2010.zip', 'w')
zip_all= ZipFile('plot(all_names)1880-2010.zip', 'w')
zip_dec= ZipFile('plot(decade_names)1880-2010.zip', 'w')

for i in range(1880,2020, 10):

    main= pd.read_csv(r"dep_counts_coha"+str(i)+".csv") 
    
    
    
#all names................................................................................................................

    #here is the selected normalized combination for agency 
    main['syn_agency']=(main['nsubj']+1)/(main['dobj']+main['nsubjpass']+2)
    
    x_female=(main.loc[main['gender']=='female'])['syn_agency'].to_list()
    
    x_male=(main.loc[main['gender']=='male'])['syn_agency'].to_list()
    
    
    
    num_bins = 20
    # the histogram of the data
    pyplot.hist(x_female, num_bins, facecolor='blue', alpha=0.5, label='female_names')
    pyplot.hist(x_male, num_bins, facecolor='green', alpha=0.5, label='male_names')
    pyplot.legend(loc='upper right')
    
    
    
    
    pyplot.xlabel('language score')
    pyplot.ylabel('counts') 
    pyplot.title(str(i)+'counts(all names)')
    # Tweak spacing to prevent clipping of ylabel
    pyplot.subplots_adjust(left=0.15)

    
       
    pyplot.savefig(str(i)+'counts(all names).png', bbox_inches="tight")
    
    
    pyplot.close()
 
    
    
    
    
    
#all names with log...................................................................................................... 
    main['syn_agency_log']=np.log10((main['nsubj']+1)/(main['dobj']+main['nsubjpass']+2))
    
    
    x_female=(main.loc[main['gender']=='female'])['syn_agency_log'].to_list()
    
    x_male=(main.loc[main['gender']=='male'])['syn_agency_log'].to_list()
    
    
    
    num_bins = 20
    # the histogram of the data
    pyplot.hist(x_female, num_bins, facecolor='blue', alpha=0.5, label='female_names')
    pyplot.hist(x_male, num_bins, facecolor='green', alpha=0.5, label='male_names')
    pyplot.legend(loc='upper right')
    
    
    
    
    pyplot.xlabel('language score')
    pyplot.ylabel('counts') 
    pyplot.title(str(i)+'counts(all names)(log of the combination)')
    # Tweak spacing to prevent clipping of ylabel
    pyplot.subplots_adjust(left=0.15)

    
    
    
    pyplot.savefig(str(i)+'counts(all names)(log of the combination).png', bbox_inches="tight")
 

    pyplot.close()




#decade names...........................................................................................................
    
    Main=main.loc[main['in_decade']==True]
    Main['syn_agency_dec']=(Main['nsubj']+1)/(Main['dobj']+Main['nsubjpass']+2)
    
    
    x_female=(Main.loc[Main['gender']=='female'])['syn_agency_dec'].to_list()
    x_male=(Main.loc[Main['gender']=='male'])['syn_agency_dec'].to_list()
    
    
    
    num_bins = 20
    # the histogram of the data
    pyplot.hist(x_female, num_bins, facecolor='blue', alpha=0.5, label='female_names')
    pyplot.hist(x_male, num_bins, facecolor='green', alpha=0.5, label='male_names')
    pyplot.legend(loc='upper right')
    pyplot.xlabel('language score')
    pyplot.ylabel('counts') 
    pyplot.title(str(i)+'counts(decade names)')
    # Tweak spacing to prevent clipping of ylabel
    pyplot.subplots_adjust(left=0.15)
    
    
    pyplot.savefig(str(i)+'counts(decade names).png', bbox_inches="tight")
    
    pyplot.close()
    
#decade names log...........................................................................................................  
    
    Main=main.loc[main['in_decade']==True]
    Main['syn_agency_dec_log']=np.log10((Main['nsubj']+1)/(Main['dobj']+Main['nsubjpass']+2))
     

    
    x_female=(Main.loc[Main['gender']=='female'])['syn_agency_dec_log'].to_list()
    x_male=(Main.loc[Main['gender']=='male'])['syn_agency_dec_log'].to_list()
    
    
    
    num_bins = 20
    # the histogram of the data
    pyplot.hist(x_female, num_bins, facecolor='blue', alpha=0.5, label='female_names')
    pyplot.hist(x_male, num_bins, facecolor='green', alpha=0.5, label='male_names')
    pyplot.legend(loc='upper right')
    pyplot.xlabel('language score')
    pyplot.ylabel('counts') 
    pyplot.title(str(i)+'counts(decade names)(log of the combination)')
    # Tweak spacing to prevent clipping of ylabel
    pyplot.subplots_adjust(left=0.15)
    
    

    
    pyplot.savefig(str(i)+'counts(decade names)(log of the combination).png', bbox_inches="tight")
    
    pyplot.close()
    
   

    zip_all_log.write(str(i)+'counts(all names)(log of the combination).png')
    zip_dec_log.write(str(i)+'counts(decade names)(log of the combination).png')
    zip_all.write(str(i)+'counts(all names).png')
    zip_dec.write(str(i)+'counts(decade names).png')
    
    
zip_all_log.close()
zip_all.close()
zip_dec_log.close()
zip_dec.close()



```

```python
import re
urls = [str("Top names of the "+ str(i) + "s.html") for i in range(1880, 2020, 10)]
decade=1880
for url in urls :
    
    male, female = [], []
    with open(url) as f:
        for line in f:
            names = re.findall("<td ?>(.*?)</td ?>", line)
            if names:
                if len(names) == 4:
                    male.append(names[0])
                    female.append(names[2])

    with open("english_names"+str(decade) +".txt", "w") as outfile:
        for m, f in zip(male, female):
            outfile.write(m + "\t" + f + "\n")

    decade+=10
```

```python
import numpy as np
import pandas as pd
from zipfile import ZipFile
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

female_names={str(j):[] for j in range(1880,2020, 10)}
male_names={str(j):[] for j in range(1880,2020, 10)}

for i in range(1880,2020, 10):

    with open ("english_names"+str(i)+".txt") as f :
        for z in f:
            female_names[str(i)].append(z.strip().split()[0].lower().strip())
            male_names[str(i)].append(z.strip().split()[1].lower().strip())




 
```

```python
all_female_names=[]
all_male_names=[]
for i in range(1880,2020, 10):
    all_female_names+=female_names[str(i)]
    all_male_names+=male_names[str(i)]
    
all_male_names=list(set(all_male_names))
all_female_names=list(set(all_female_names))
```

```python
set.intersection(set(all_female_names), set(all_male_names))
```

```python
overlap=list(set.intersection(set(all_female_names), set(all_male_names)))



for i in range(1880,2020, 10):
    
    main= pd.read_csv("dep_counts_coha"+str(i)+".csv") 
    words=main['word'].to_list()
    in_decade=[]
    gender=[]
    for word in words:
        if word in (female_names[str(i)]+male_names[str(i)]):
            in_decade.append(True)
        if word not in list(female_names[str(i)]+male_names[str(i)]):
            in_decade.append(False)
        if word in overlap :
            gender.append('both')     
        if (word in all_female_names)and(word not in overlap) :
            gender.append('female')
        if (word in all_male_names)and(word not in overlap) :  
            gender.append('male')
        if word not in (all_female_names+all_male_names):
            gender.append(None)
    main['gender']=gender        
    main['in_decade']=in_decade     
    main.to_csv(r'dep_counts_coha'+str(i)+'.csv')
```
